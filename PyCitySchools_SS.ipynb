```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
#school_data_to_load = "Resources/schools_complete.csv"
#student_data_to_load = "Resources/studentss_complete.csv"
school_data_to_load="C:/Users/SS3/Desktop/GTBC/GTBC_HW/04-Pandas/PyCitySchools/Resources/schools_complete.csv"
student_data_to_load = "C:/Users/SS3/Desktop/GTBC/GTBC_HW/04-Pandas/PyCitySchools/Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

#print(school_data.head())
student_data.head()

SchoolCols = school_data.columns
StudentCols = student_data.columns
print(SchoolCols)
print(StudentCols)


student_data2 = student_data.rename(columns={'Student ID': 'Student_ID','name': 'Student_Name','gender':'gender','grade':'grade','school':'School_Name','reading_score':'reading_score',
                             'math_score':'math_score'}, inplace=False)
student_data.head()
school_data2 = school_data.rename(columns={'School ID':'School_ID','name':'School_Name','type':'type', 'size':'size', 'budget':'budget'},inplace=False)
#school_data.head()

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data2, school_data2, how="left", on=["School_Name"])

school_data_complete.head()
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)
  
* Create a dataframe to hold the above results

```python



df_School_Summ = school_data2.drop(columns=['School_ID'])

df_School_Summ['Per Student Budget'] = df_School_Summ['budget']/df_School_Summ['size']


stud_avg_grpd = school_data_complete.groupby('School_Name')['math_score','reading_score'].mean()
df_Std_avg_grpd = pd.DataFrame(stud_avg_grpd)
df_School_Summ = pd.merge(df_School_Summ, df_Std_avg_grpd, how="inner", on=["School_Name"])
#print(df_School_Summ)




df_stud_mathpass = school_data_complete.loc[(school_data_complete['math_score']>70)]
stud_mathpass_grpd = df_stud_mathpass.groupby('School_Name')['Student_ID'].count()
df_stud_mathpass_grpd = pd.DataFrame(stud_mathpass_grpd)
df_stud_mathpass_grpd = df_stud_mathpass_grpd.rename(columns={'Student_ID':'MathPass_Cnt'})

df_stud_rdpass = school_data_complete.loc[(school_data_complete['reading_score']>70)]
stud_rdpass_grpd = df_stud_rdpass.groupby('School_Name')['Student_ID'].count()
df_stud_rdpass_grpd = pd.DataFrame(stud_rdpass_grpd)
df_stud_rdpass_grpd = df_stud_rdpass_grpd.rename(columns={'Student_ID':'ReadPass_Cnt'})
#print(df_stud_rdpass_grpd)



df_School_Summ = pd.merge(df_School_Summ, df_stud_mathpass_grpd, how="inner", on=["School_Name"])
df_School_Summ = pd.merge(df_School_Summ, df_stud_rdpass_grpd, how="inner", on=["School_Name"])


df_School_Summ['% Passing Math'] = df_School_Summ['MathPass_Cnt']/df_School_Summ['size']

df_School_Summ['% Passing Reading'] = df_School_Summ['ReadPass_Cnt']/df_School_Summ['size']

df_School_Summ['% Overall Passing Rate'] = ((df_School_Summ['% Passing Math']+df_School_Summ['% Passing Reading'])/2)

df_School_Summ = df_School_Summ.drop(columns=['MathPass_Cnt','ReadPass_Cnt'])

df_School_Summ_cln = df_School_Summ.rename(columns={'type':'School Type','size':'Total Students'
                                                    ,'budget':'Total School Budget','math_score':'Average Math Score'
                                                    ,'reading_score':'Average Reading Score'
                                                    ,'School_Name':''})

df_School_Summ_cln = df_School_Summ_cln.set_index('')
#df_School_Summ_cln2.reset_index(drop = False, inplace = True)
df_School_Summ_cln


```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
# Sample bins. Feel free to create your own bins.
spending_bins = [0, 585, 615, 645, 675]
group_names = ["<$585", "$585-615", "$615-645", "$645-675"]
```

```python
df_School_Summ_sub = df_School_Summ_cln.drop(columns=['Total School Budget','Total Students','% Overall Passing Rate'])

df_School_Summ_sub['Spending Ranges (Per Student)'] = pd.cut(df_School_Summ_sub['Per Student Budget'], bins=spending_bins, labels=group_names)
df_School_Summ_grp = df_School_Summ_sub.groupby('Spending Ranges (Per Student)')
df_School_Summ_PSB = df_School_Summ_grp.mean()
df_School_Summ_PSB['% Overall Passing Rate']=(df_School_Summ_PSB['% Passing Math']+df_School_Summ_PSB['% Passing Reading'])/2
df_School_Summ_PSB_cln = df_School_Summ_PSB.drop(columns=['Per Student Budget'])
df_School_Summ_PSB_cln
```
